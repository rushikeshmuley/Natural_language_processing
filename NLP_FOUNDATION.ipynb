```python
import nltk
```

```python
nltk.download("all")
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
example = """ Narendra Damodardas Modi; born 17 September 1950)[b] is an Indian politician serving as the 14th and current prime minister of India since 2014. Modi was the chief minister of Gujarat from 2001 to 2014 and is the Member of Parliament from Varanasi. He is a member of the Bharatiya Janata Party (BJP) and of the Rashtriya Swayamsevak Sangh (RSS), a right-wing Hindu nationalist paramilitary volunteer organisation. He is the longest serving prime minister from outside the Indian National Congress.
"""
```

```python
sent_tokenize(example)
```

```python
for sent in sent_tokenize(example):
  print(sent)
```

```python
word_tokenize(example)
```

```python
for word in word_tokenize(example):
  print(word)
```

```python
# whitespace tokenizer
```

```python
white_token = nltk.tokenize.WhitespaceTokenizer()
```

```python
white_token.tokenize(example)
```

```python
for white in white_token.tokenize(example):
  print(white)
```

```python
text='I\m with you till your selection in DataScience in U.K.!'
```

```python
text
```

```python
# Regular Expression
```

```python
import re
words=re.split(r'\W+',text)
print(words)
```

```python
# Compile all together
import re
import string
re_puct=re.compile('[%s]' %re.escape(string.punctuation))
re_puct
original =[re_puct.sub('',w) for w in words]
print(original)
```

```python
re_puct=re.compile('[^%s]' %re.escape(string.printable))
re_puct
original =[re_puct.sub('',w) for w in words]
print(original)
```
